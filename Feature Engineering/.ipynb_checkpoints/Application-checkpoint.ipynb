```python
import os
import numpy as np
import pandas as pd
from ggplot import *
import matplotlib.pyplot as plt
import seaborn as sns
import copy
import plotly.plotly as py
import plotly.graph_objs as go
from sklearn import preprocessing

pd.options.display.max_columns = 200
```

```python
# 缺失率统计
def find_missing(data):
    # number of missing values
    count_missing = data.isnull().sum().values
    # total records
    total = data.shape[0]
    # percentage of missing
    ratio_missing = count_missing / total
    # return a dataframe to show: feature name, # of missing and % of missing
    return pd.DataFrame(data={'missing_count': count_missing, 'missing_ratio': ratio_missing}, index=data.columns.values)

# 类别型变量-分布    
def plot_categorical(data, x, size=[8, 4], xlabel_angle=0, title=''):
    '''use this for ploting the count of categorical features'''
    temp = data[x]
    temp[temp.isnull()] = "Miss"
    
    plotdata = temp.value_counts()
    plt.figure(figsize=size)
    sns.barplot(x=plotdata.index, y=plotdata.values)
    plt.title(title)
    if xlabel_angle != 0:
        plt.xticks(rotation=xlabel_angle)   # 轴坐标刻度-角度
    plt.show()
    
# 类别特征：条件分布情况
def plot_categorical_bylabel(data, x, by_x, by_class=[0,1], size=[12 ,6], xlabel_angle=0):
    '''use it to compare the distribution between label 1 and label 0'''
    plot_df = data[[by_x, x]]
    plot_df.loc[plot_df[x].isnull(),x] = "Miss"
    
    plt.figure(figsize = size)
    l0 = plot_df.loc[plot_df[by_x]==by_class[0], x].value_counts(normalize = True).mul(100)
    l1 = plot_df.loc[plot_df[by_x]==by_class[1], x].value_counts(normalize = True).mul(100)
    
    if (any(l1.index == "Miss") & any(l0.index == "Miss")):
        print(l1.loc['Miss']-l0.loc['Miss'])
    
    plt.subplot(1,2,1)
    sns.barplot(x = l1.index, y=l1.values)
    plt.title(by_class[1] + ":" + x)
    plt.xticks(rotation=xlabel_angle)
    plt.subplot(1,2,2)
    sns.barplot(x = l0.index, y=l0.values)
    plt.title(by_class[0] + ":" + x)
    plt.xticks(rotation=xlabel_angle)
    plt.show()

# 填充 Missing
# 如果正负样本 Miss占比相差>=5%,则缺失作为一类
def missing_fill(data, x):
    df = data[['TARGET', x]]
    df.loc[df[x].isnull(),x] = "Miss"
    
    if (any(df[x].unique()=="Miss")):
        l1 = df.loc[df.TARGET==1, x].value_counts(normalize = True).mul(100)
        l0 = df.loc[df.TARGET==0, x].value_counts(normalize = True).mul(100)
        
        if (abs(l1.loc['Miss']-l0.loc['Miss'])>=0.05):
            temp = copy.copy(data[x])
            temp.loc[temp.isnull()] = "Miss"
            data['temp'] = temp
        else:
            temp = copy.copy(data[x])
            temp.loc[temp.isnull()] = data[x].mode().values
            data['temp'] = temp
    
        data = data.drop([x], axis=1)
        data = data.rename(columns={"temp": x})
    
    return(data)

# LabelEncoding
def labelencoding(data, x):
    le = preprocessing.LabelEncoder()
    le.fit(data[x].tolist())
    print(le.classes_)
    
    temp = le.transform(app_all[x])
    data['temp'] = temp
    
    data = data.drop([x], axis = 1)
    data.rename(columns={"temp": x}, inplace = True)
    return data
```

```python
# 读取数据
app_train = pd.read_csv("../../Data/application_train.csv")
app_test = pd.read_csv("../../Data/application_test.csv")
app_all = pd.read_csv("../../Data_/app_train_test.csv")

app_train_columns = app_train.columns.values

# 选出类别型变量
cato_features = app_all.select_dtypes(include=['object']).columns

app_train_columns = app_train.columns.values
app_all = app_all[app_train_columns]
app_all['TID'] = np.arange(app_all.shape[0])
app_all['Set'] = np.where(app_all['TARGET'].isnull(),"Test","Train")

print(app_all.shape)
```

```python
# 特征缺失情况统计
miss_sta_all = find_missing(app_all).sort_values(by = "missing_ratio", ascending = False)
miss_sta_train = find_missing(app_train).sort_values(by = "missing_ratio", ascending = False)
miss_sta_test = find_missing(app_test).sort_values(by = "missing_ratio", ascending = False)

miss_sta = pd.concat([miss_sta_all,miss_sta_train,miss_sta_test], axis = 1)
miss_sta['train_test_diff'] = miss_sta.iloc[:,3] - miss_sta.iloc[:,5]
miss_sta
miss_sta.loc[miss_sta.train_test_diff>=0.03,].index
```
